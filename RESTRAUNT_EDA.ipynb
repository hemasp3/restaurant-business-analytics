```python
#no.of invoices bill above overall avg bill
invgroup=data.groupby('Invoice No.')['Final Total'].sum()
#average amount spend
invgroupavg=invgroup.mean()
print("Avg amount spend: ",invgroupavg)
#bill>avg bill
ans=invgroup[invgroup>invgroupavg].count()
print("Invoices spend above avg: ",ans)
```

```python
#top selling Item on each date
qty=data.groupby(['Date','Item Name'])['Qty.'].sum().reset_index()
top=(qty.sort_values(['Date','Qty.'],ascending=[True,False]).groupby('Date').first().reset_index())
top
```

```python
#analysing do high quantity items always generate high revenue
revbyqty=(data.groupby('Item Name').agg(total_quantity=('Qty.','sum'),total_revenue=('Final Total','sum')).reset_index())
ans=revbyqty[['total_quantity','total_revenue']].corr()
correlation_value = ans.loc['total_quantity', 'total_revenue']
print(correlation_value)
if(correlation_value > 0.7):
  print(" high quantity items generally generate high revenue.")
else:
  print("high quantity items do not always generate high revenue.")
```

```python
revbyqty=data.groupby('Item Name')[['Qty.','Final Total']].sum()
revbyqty['qrank']=revbyqty['Qty.'].rank(method='first',ascending=False)
revbyqty['rrank']=revbyqty['Final Total'].rank(method='first',ascending=False)
revbyqty['rdiff']=revbyqty['rrank']-revbyqty['qrank']
revbyqty.sort_values('rdiff',ascending=False)
if(revbyqty['rdiff']!=0).any():
  print("High quantity items do not always generate high revenue")
else:
  print("High quantity items always generate high revenue")
```

```python
#item that contributed highest percentage on total revenue
rev=data.groupby('Item Name')['Final Total'].sum().reset_index()
rev['contribute']=(rev['Final Total']/rev['Final Total'].sum())*100
rev.sort_values('contribute',ascending=False).head(1)
```
